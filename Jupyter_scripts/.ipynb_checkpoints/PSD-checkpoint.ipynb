```python
'''
This script performs optimization of PDS algorithms, PSD algorithms,
calculation of FOM and plotting the PSD performance graphs
'''

%reset
#!/usr/bin/env python
#-*- coding: utf-8 -*-
import sys
import os
import numpy as np
import pandas as pd
from scipy import signal
from scipy.optimize import curve_fit
import scipy.fftpack
import matplotlib.pyplot as plt

def main():
    threshold = 0.05 # trigger threshold - 5 percent of maximum of the pulse

    path = "/home/sommema4/git/SiPM_simulator/data/preproc/MicroFC60035_EJ-276"
    path_out = "/home/sommema4/git/SiPM_simulator/data/proc/MicroFC60035_EJ-276"
    make_directory(path_out)

    lst = []
    for filename in os.listdir(path):
        file_in = os.path.join(path, filename)
        if filename.endswith(".csv"):
            cut_off_freq, sample_freq = strip_filename(filename)
            cut_off_freq = float(cut_off_freq)
            sample_freq = float(sample_freq)
            if (cut_off_freq > sample_freq / 2.0): # Nyquist theorem
                continue
            print(cut_off_freq, sample_freq)
            df = read_csv(file_in)
            rows, columns = df.shape
            columns = columns - 2
            
            # Slice dataframe to the individual components of the signal
            total, light, afterpulse, crosstalk, dark_current = slice_dataframe(df, columns)

            columns = columns - 1
            # Optimize PSD techniques
            df_n, df_p = mean_dataframe(total)
            df_n = amplitude_normalization(df_n)
            df_p = amplitude_normalization(df_p)
            popt_ccm = ccm_optimization(df_n, df_p, columns, threshold, sample_freq)
            popt_fga = fga_optimization(df_n, df_p, columns)
            
            print(popt_fga)
            popt_fga = 1

            # Find out the pulse properties
            total = properties(total, columns, threshold, sample_freq)
            #print(total)
            # Perform PSD techniques
            total['ccm'] = ccm(total, popt_ccm)
            total['fga'] = fga(total, columns, popt_fga, sample_freq)
            total['caa'] = caa(total)
            #print(total)
            FOM_ccm, FOM_caa = plot_PSD(total)
            print(FOM_ccm, FOM_caa)
            lst.append((FOM_ccm, FOM_caa, cut_off_freq, sample_freq))
            
    df_fom = pd.DataFrame(lst, columns=['FOM_ccm', 'FOM_caa', 'cut_off', 'sample_freq'])
    file_out = os.path.join(path_out, 'FOM.csv')
    df_fom.to_csv(file_out)
    plot_FOM(df_fom)
        
def strip_filename(filename):
    base = os.path.basename(filename)
    name = os.path.splitext(base)[0]
    cut_off_freq, sample_freq = name.split("_")
    return cut_off_freq, sample_freq

def make_directory(dir_name):
    if not (os.path.isdir(dir_name)):
        try:
            os.makedirs(dir_name)
        except OSError:
            print ("Creation of the directory %s failed" % dir_name)
        else:
            print ("Successfully created the directory %s " % dir_name)

def read_csv(filename):
    # Read data to pandas dataframe and transpose it
    df = pd.read_csv(filename, sep=',', header=None)
    #df = df.iloc[:, :-1]
    return df

def slice_dataframe(df, columns):
    light = df.iloc[0::4, 1:columns]
    afterpulse = df.iloc[1::4, 1:columns]
    crosstalk = df.iloc[2::4, 1:columns]
    dark_current = df.iloc[3::4, 1:columns]
    energy = df.iloc[0::4, -2]
    particle = df.iloc[0::4, -1]
    energy = energy.reset_index(drop=True)
    particle = particle.reset_index(drop=True)
    total = pd.DataFrame(light.values + crosstalk.values + afterpulse.values + dark_current.values)

    total = total.reset_index(drop=True)
    total['energy'] = energy
    total['particle'] = particle

    return total, light, afterpulse, crosstalk, dark_current

def amplitude_normalization(x):
    return x / x.max()

def mean_dataframe(df):
    df = df.drop(columns=['energy'])
    df_n = df[df['particle']=='n']
    df_p = df[df['particle']=='p']
    df_n = df_n.mean(axis=0)
    df_p = df_p.mean(axis=0)
    #print(df_n, df_p)
    return df_n, df_p

def ccm_optimization(df_n, df_p, columns, threshold, sample_freq):
    lst_out, start, end = [], [], []
    for i in [df_n, df_p]:
        max_index = i.idxmax()
        temp_end = np.abs(i.iloc[max_index:columns] - threshold)
        pulse_end = max_index + temp_end.argmin()
        temp_beginning = np.abs(i.iloc[:max_index] - threshold)
        pulse_beginning = temp_beginning.argmin()
        start.append(pulse_beginning)
        end.append(pulse_end)
        lst_out.append(pulse_end - max_index)
        
    temp = []
    n_integral = integrate(df_n.iloc[:columns], start[0], end[0])
    p_integral = integrate(df_p.iloc[:columns], start[1], end[1])
    for i in range(min(lst_out)):
        n_tail = integrate(df_n.iloc[:columns], df_n.idxmax() + i, end[0])
        p_tail = integrate(df_p.iloc[:columns], df_p.idxmax() + i, end[1])
        temp.append(np.abs((n_tail / n_integral) - (p_tail / p_integral)))
        
    temp = np.array(temp)
    out = temp.argmax()
    #print(out)
    return out

def fga_optimization(df_n, df_p, columns):
    fft_n = fft_pulse(df_n.iloc[:columns])
    fft_p = fft_pulse(df_p.iloc[:columns])
    temp = []
    for i in range(len(fft_n)):
        fga_n = fft_n[0] - fft_n[i]
        fga_p = fft_p[0] - fft_p[i]
        temp.append(np.abs(fga_n - fga_p))
        
    temp = np.array(temp)
    out = temp.argmax()
    #print(out)
    return out

def properties(df, columns, threshold, sample_freq):
    T = 1 / sample_freq
    #print(df.iloc[:, :columns])
    df['amplitude'] = df.iloc[:, :columns].max(axis=1)
    df['idxmax'] = df.iloc[:, :columns].idxmax(axis=1)
    df_norm = df.iloc[:, :columns].apply(amplitude_normalization, axis=1)
    df_norm['idxmax'] = df['idxmax']
    #print('maximum:', df_norm.iloc[:, :columns].max(axis=1))
    df['start'] = df_norm.apply(start_pulse, args=(columns, threshold), axis=1)
    df['stop'] = df_norm.apply(stop_pulse, args=(columns, threshold), axis=1)
    df['pulse_duration'] = (df['stop'] - df['start']) * T
    df['integral'] = df.apply(integral_pulse, axis=1)
    #print(df)
    return df

def start_pulse(x, columns, threshold):
    max_index = int(x['idxmax'])
    #print(max_index, x.iloc[:max_index])
    temp_beginning = np.abs(x.iloc[:max_index] - threshold)
    temp_beginning = np.array(temp_beginning)
    pulse_beginning = temp_beginning.argmin()
    return pulse_beginning

def stop_pulse(x, columns, threshold):
    max_index = int(x['idxmax'])
    temp_end = np.abs(x[max_index:columns] - threshold)
    temp_end = np.array(temp_end)
    pulse_end = temp_end.argmin()
    #print(max_index, columns, max_index + pulse_end)
    return max_index + pulse_end

def integral_pulse(x):
    start = x['start']
    stop = x['stop']
    return x[start:stop].sum()

def integrate(x, start, stop):
    return x[start:stop].sum()

def ccm_pulse(x, popt):
    #print(popt)
    start = x['idxmax'] + popt
    stop = x['stop']
    integral = x['integral']
    tail = integrate(x, start, stop)
    #print(start, stop, tail, integral)
    return 1 - (tail / integral)

def ccm(df, popt):
    ccm = df.apply(ccm_pulse, args=([popt]), axis=1)
    return ccm

def caa(df):
    return df['amplitude'] / df['integral']

def fft_pulse(x):
    ps = pd.Series(np.abs(np.fft.fft(x))**2)
    return ps
    
def fft(df):
    df_fft = df.apply(fft_pulse, axis=1)
    return df_fft

def fga_pulse(x, popt):
    out = pd.Series(x[0] - x[popt])
    #print(type(out))
    return out

def fga(df, columns, popt, sample_freq):
    freq = popt * (1 / sample_freq) # ?
    df_fft = fft(df.iloc[:, :columns])
    #print(df_fft)
    fga = df_fft.apply(fga_pulse, args=[popt], axis=1)
    fga = fga / freq # normalization
    return fga

def GMM(x, c1, mu1, sigma1, c2, mu2, sigma2):
    #print(c1, mu1, sigma1, c2, mu2, sigma2)
    res = c1 * np.exp(-(x - mu1)**2.0 / (2.0 * sigma1**2.0)) + c2 * np.exp(-(x - mu2)**2.0 / (2.0 * sigma2**2.0))
    return res

def fit_GMM(x, c1, mu1, sigma1, c2, mu2, sigma2):
    fit = GMM(x, c1, mu1, sigma1, c2, mu2, sigma2)
    return fit

def gauss(x, a, x0, sigma):
    return a * np.exp(-(x - x0)**2 / (2 * sigma**2))
    
def borders(x, edge=0.1):
    delta = edge * (x.max() - x.min())
    return x.min() - delta, x.max() + delta

def find_GMM_params(df, tag, tag_x):
    bins_int = len(tag_x)
    i = bins_int
    params_condition = False
    params = np.zeros(6)
    explanation = ['amplitude of the first peak: ', 'position of the first peak: ', 'sigma of the first peak: ', 'amplitude of the second peak: ', 'position of the second peak: ', 'sigma of the second peak: ']
    
    while (i > 9):
        temp_hist, temp_bins = np.histogram(df[tag], bins=i)
        maximum = np.max(temp_hist)
        peak_pos, peak_height = signal.find_peaks(temp_hist, height=maximum//10, threshold=None, distance=bins_int//5)
        peak_height = peak_height['peak_heights']
        #print(peak_pos, peak_height)
        if (len(peak_pos) == 2):
            ratio = (float(i) / float(bins_int))
            maxima = [ratio * peak_height[0], ratio * peak_height[1]]
            position = [temp_bins[peak_pos[0]], temp_bins[peak_pos[1]]]
            sigma = [(tag_x[-1] - tag_x[0]) / 20, (tag_x[-1] - tag_x[0]) / 10]
            '''
            print(temp_hist)
            print('maxima: ', maxima)
            print('position: ', position)
            print('sigma: ', sigma)
            '''
            params_condition = True
            params = [maxima[0], position[0], sigma[0], maxima[1], position[1], sigma[1]]
            break
        i = i - 1
    
    if (params_condition == False): # if the algorithm failed to guess parameters than write them down yourself
        print(tag, 'algorithm failed')
        plt.hist(df[tag], bins=tag_x)
        plt.pause(1)
        for i, sign in enumerate(explanation):
            params[i] = input(sign)
    return params

def calculate_FOM(fit):
    c1, mu1, sigma1, c2, mu2, sigma2 = fit
    #print(c1, mu1, sigma1, c2, mu2, sigma2)
    FWHM_1 = 2.35482 * sigma1
    FWHM_2 = 2.35482 * sigma2
    FOM = (np.abs(mu1 - mu2)) / (FWHM_1 + FWHM_2)
    return FOM
    
def plot_PSD(df):
    bins_int = 50
    ccm_min, ccm_max = borders(df['ccm'])
    fga_min, fga_max = borders(df['fga'])
    caa_min, caa_max = borders(df['caa'])
    
    ccm_x = np.linspace(ccm_min, ccm_max, num=bins_int)
    fga_x = np.linspace(fga_min, fga_max, num=bins_int)
    caa_x = np.linspace(caa_min, caa_max, num=bins_int)
    
    hist_ccm, _ = np.histogram(df['ccm'], bins=ccm_x)
    hist_fga, _ = np.histogram(df['fga'], bins=fga_x)
    hist_caa, _ = np.histogram(df['caa'], bins=caa_x)
    
    params_ccm = find_GMM_params(df, 'ccm', ccm_x)
    params_caa = find_GMM_params(df, 'caa', caa_x)
    
    try:
        fit_ccm = curve_fit(fit_GMM, xdata=ccm_x[:-1], ydata=hist_ccm, p0=params_ccm, maxfev = 1400)[0]
    except:
        fit_ccm = [1, 1, 1, 1, 1, 1]
    try:
        fit_caa = curve_fit(fit_GMM, xdata=caa_x[:-1], ydata=hist_caa, p0=params_caa, maxfev = 1400)[0]
    except:
        fit_caa = [1, 1, 1, 1, 1, 1]
    
    FOM_ccm = calculate_FOM(fit_ccm)
    FOM_caa = calculate_FOM(fit_caa)
    
    df_n = df[df['particle']=='n']
    df_p = df[df['particle']=='p']
    
    #print(df_n)
    #print(df_p)
    
    fig = plt.figure(figsize=(18,9))
    ax1 = fig.add_subplot(3,2,1)
    ax1.plot(df_n['amplitude'], df_n['ccm'], '+', color='r')
    ax1.plot(df_p['amplitude'], df_p['ccm'], '+', color='b')
    ax1.set_ylabel('ccm [-]')
    ax1.set_ylim(ccm_min, ccm_max)
    #ax1.legend(loc='upper right')
    
    ax4 = fig.add_subplot(3,2,2)
    ax4.step(hist_ccm, ccm_x[:-1])
    ax4.plot(GMM(ccm_x, *fit_ccm), ccm_x, '-')
    ax4.set_ylabel('ccm [-]')
    ax4.set_ylim(ccm_min, ccm_max)
    #ax4.legend(loc='upper right')
    
    ax2 = fig.add_subplot(3,2,3)
    ax2.plot(df_n['amplitude'], df_n['fga'], '+', color='r')
    ax2.plot(df_p['amplitude'], df_p['fga'], '+', color='b')
    ax2.set_ylabel('fga [-]')
    ax2.set_ylim(fga_min, fga_max)
    #ax2.legend(loc='upper right')
    
    ax3 = fig.add_subplot(3,2,5)
    ax3.plot(df_n['amplitude'], df_n['caa'], '+', color='r')
    ax3.plot(df_p['amplitude'], df_p['caa'], '+', color='b')
    ax3.set_ylim(caa_min, caa_max)
    ax3.set_xlabel('Amplitude [A]')
    ax3.set_ylabel('caa [-]')
    #ax3.legend(loc='upper right')
    
    ax6 = fig.add_subplot(3,2,6)
    ax6.step(hist_caa, caa_x[:-1])
    ax6.plot(GMM(caa_x, *fit_caa), caa_x, '-')
    ax6.set_ylabel('caa [-]')
    ax6.set_ylim(caa_min, caa_max)
    #ax6.legend(loc='upper right')
    
    plt.pause(1)
    plt.close()
    
    return FOM_ccm, FOM_caa

def plot_FOM(df):
    color_pie = ['r', 'g', 'b', 'k', 'y']
    fig = plt.figure(figsize=(18,9))
    ax1 = fig.add_subplot(2,1,1)
    ax2 = fig.add_subplot(2,1,2)
    ax1.set_xscale('log')
    ax2.set_xscale('log')
    
    sample = np.unique(df['sample_freq'])
    for i, freq in enumerate(sample):
        df_freq = df[df['sample_freq']==freq]
        df_sort = df_freq.sort_values(by=['cut_off'])
        print(df_sort)
        ax1.plot(df_sort['cut_off'], df_sort['FOM_ccm'], color=color_pie[i], label=str(freq)+' MHz')
        ax2.plot(df_sort['cut_off'], df_sort['FOM_caa'], color=color_pie[i], label=str(freq)+' MHz')
        
    plt.pause(3)

        
main()
```
